```python
import numpy as np
import pandas as pd
from numpy import random
ds=pd.read_csv("/content/drive/MyDrive/Iris.csv")
print(ds.head(13))
```

```python
#now splitting the datset into features and labels
features=ds.values[:,1:-1] #taking features as an array with all rows and two feature cols from the original dataset
print(features)


# now for labels or classes
lables=ds.values[:,-1] # taking only label col as an array
print(lables)
```

```python
# manual splitting
m = len(features)
train_size = 100
test_size = m-train_size
idx = np.random.randint(0,m,train_size)
train_x = features[idx]
train_y = lables[idx]
idx = np.random.randint(0,m,test_size)
test_x = features[idx]
test_y = lables[idx]
print(train_x.shape)
print(test_x.shape)
```

```python
# def euclidean_distance(a,b):
#   return np.sqrt(np.sum((a-b)**2))


#using Minkowski distance function we can drive euclidean distance when p =2 and hamanttan distance when p=1, 
def MinkowskiGeneralisation(a,b,p=2):
    po= np.sum(np.abs(a-b)**p)
    return np.power(po,1/p)





         
 
```

```python
def get_distances(test_ins,features):
   m = features.shape[0]
   distances = np.zeros(m)
   for i in range(m):
          #  d = euclidean_distance(test_ins, features[i])
            d = MinkowskiGeneralisation(test_ins, features[i])
            distances[i]=d
   return distances
```

```python
def accuracy_score(pred_y):
  print(pred_y==test_y)
  return (np.sum(pred_y == test_y)/ len(pred_y)) * 100
```

```python
k = 3
num_classes = 3
pred_y = np.zeros(len(test_y))
# this loop is for each test instance

for i in range(len(test_x)):
   Frequency= np.zeros(num_classes)     
   label=0                              #for each test instance their will be a different label set so reinitializing the label
   test_ins= np.array(test_x[i])
   print(test_ins)
   distances = get_distances(test_ins,train_x)
   sorted_dist = np.argsort(distances)
   print(sorted_dist)
   # print(sorted_dist)
   for ii in range(k): # this loop is for majority voting\
      label =int( train_y[sorted_dist[ii]])      # label will contain a set of labels for each dataset
      print(label)
      Frequency[label]= Frequency[label]+1       # here majority voting is done that is in k nearest neighnors which category is in higher number
   print(Frequency)
   idx = np.argmax(Frequency)     #temporary storage of  predicted category in idx 
   print(f'index: {idx}')
   pred_y[i]= idx                  #for each i-test instance we are storing its predicted category in i-predicted array 

















# print(pred_y)
acc = accuracy_score(pred_y)
print(f'accuracy of KNN classifier on iris dataset  train-test split ratio {len(train_x)/len(test_x)} is {acc}')
```

```python
#calculating sample variance 
def SampleVariance(ar,Mean,size):
  sampleVar=0
  for i in range(size):
    sampleVar+=np.square(ar[i]-Mean)
  return sampleVar/(size-1)



#calculate covariance
def Covariance(arr1,arr2,Mean1,Mean2,lenn):
  covariancee=0
  for i in range(lenn):
     covariancee+=((arr1[i]-Mean1)*(arr2[i]-Mean2))
  return covariancee/(lenn-1)







import array as arr
a = arr.array('i', [1692, 1978, 1884, 2151, 2519])
b = arr.array('i', [68, 102, 110, 112, 154])

Mean_x=0
Mean_y=0
z=0
y=0
#calculating Mean of a and b
for i in range(len(a)):
  z+=a[i]
  y+=b[i]

Mean_x=z/len(a)
Mean_y=y/len(b)
print(f'Mean of a and b is :{Mean_x}{Mean_y}')

#now calculating sample Variance for a and b
SampleVarianceeOfa = SampleVariance(a,Mean_x,len(a))

SampleVarianceeOfb = SampleVariance(b,Mean_y,len(b))

print(f'Sample variance of a and b is  : {SampleVarianceeOfa} {SampleVarianceeOfb} ')






#calculating Covariance of a and b
Cov= Covariance(a,b,Mean_x,Mean_y,len(a))
print(f'Covariance of a and b is :{Cov}')
```
